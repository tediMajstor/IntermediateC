Klasa **Token** predstavlja uređeni par (klasa, leksema).

Medota **str** vraća string reprezentaciju tokena koja se koristi u procesu pronalaženja grešaka.

```
class Token:
    def __init__(self, class_, lexeme):
        self.class_ = class_
        self.lexeme = lexeme

    def __str__(self):
        return "<{} {}>".format(self.class_, self.lexeme)
```

```
class Lexer:

    def die_type(self, expected, found):
        self.die("Expected: {}, Found: {}".format(expected, found))

    def __init__(self, text):
        self.text = text
        self.len = len(text)
        self.pos = -1
        self.tokens=[]

    def read_space(self):
        while self.pos + 1 < self.len and self.text[self.pos + 1].isspace():
            self.next_char()

    def read_number(self):
        lexeme = self.text[self.pos]
        while self.pos + 1 < self.len and (self.text[self.pos + 1].isdigit() or (self.text[self.pos + 1]=='.' and self.text[self.pos+2]!='.')):
            lexeme += self.next_char()
        if '.' in lexeme:
          print(f". in {lexeme}")
          return float(lexeme)
        return int(lexeme)

    def read_char(self):
        self.pos += 1
        lexeme = self.text[self.pos]
        self.pos += 1
        return lexeme

    def read_string(self):
        lexeme = ''
        while self.pos + 1 < self.len and self.text[self.pos + 1] != '\'':
            lexeme += self.next_char()
        self.pos += 1
        return lexeme

    def read_keyword(self):
        lexeme = self.text[self.pos]
        while self.pos + 1 < self.len and (self.text[self.pos + 1].isalnum() or self.text[self.pos + 1]=='_'):
            lexeme += self.next_char()
        if lexeme == 'if':
            return Token(Class.IF, lexeme)
        elif lexeme == 'then':
            return Token(Class.THEN, lexeme)
        elif lexeme == 'else':
            return Token(Class.ELSE, lexeme)
        elif lexeme == 'while':
            return Token(Class.WHILE, lexeme)
        elif lexeme == 'repeat':
            return Token(Class.REPEAT,lexeme)
        elif lexeme == 'until':
            return Token(Class.UNTIL,'until')
        elif lexeme == 'for':
            return Token(Class.FOR, lexeme)
        elif lexeme == 'to':
            return Token(Class.TO, lexeme)
        elif lexeme == 'downto':
            return Token(Class.DOWNTO, lexeme)
        elif lexeme == 'do':
            return Token(Class.DO, lexeme)
        elif lexeme == 'break':
            return Token(Class.BREAK, lexeme)
        elif lexeme == 'continue':
            return Token(Class.CONTINUE, lexeme)
        elif lexeme == 'exit':      #exit umesto return
            return Token(Class.EXIT, lexeme)
        elif lexeme=='integer' or lexeme=='char' or lexeme=='procedure' or lexeme=='function' or lexeme=='string' or lexeme=='real' or lexeme=='boolean':   #umesto void
            return Token(Class.TYPE, lexeme)
        elif lexeme.lower() == 'array':return Token(Class.ARRAY,lexeme)
        elif lexeme.lower() == 'var': return Token(Class.VAR, 'var')
        elif lexeme.lower() == 'procedure':return Token(Class.TYPE,lexeme)
        elif lexeme.lower() == 'function':return Token(Class.TYPE,lexeme)
        elif lexeme.lower() == 'of': return Token(Class.OF, lexeme)
        elif lexeme.lower() == 'mod': return Token(Class.MOD, lexeme)
        elif lexeme.lower() == 'div': return Token(Class.DIV, lexeme)
        elif lexeme.lower() == 'and': return Token(Class.AND, 'AND')
        elif lexeme.lower() == 'or': return Token(Class.OR, 'OR')
        elif lexeme.lower() == 'not': return Token(Class.NOT, 'NOT')
        elif lexeme.lower() == 'begin': return Token(Class.BEGIN, 'begin')
        elif lexeme.lower() == 'false': return Token(Class.FALSE, 'false')
        elif lexeme.lower() == 'true': return Token(Class.TRUE, 'true')
        elif lexeme.lower() == 'end':
          if self.text[self.pos+1] == '.':
            self.next_char()
            return Token(Class.FEND, 'end.')
          else:return Token(Class.END, 'end')

        return Token(Class.ID, lexeme)

    def next_char(self):
        self.pos += 1
        if self.pos >= self.len:
            return None
        return self.text[self.pos]

    def next_token(self):
        self.read_space()
        curr = self.next_char()
        if curr is None:
            return Token(Class.EOF, curr)
        token = None

        if curr.isalpha():
            token = self.read_keyword()

        elif curr.isdigit():
            num=self.read_number()
            if type(num).__name__== 'int':
              token = Token(Class.INTEGER,num)
            elif type(num).__name__== 'float':
              token = Token(Class.FLOAT, num)

        elif curr=='-':
          if self.pos + 1 < self.len and self.text[self.pos+1].isdigit():
            num=self.read_number()
            if type(num).__name__== 'int':
              token = Token(Class.INTEGER,num)
            elif type(num).__name__== 'float':
              token = Token(Class.FLOAT, num)
          else:
            token = Token(Class.MINUS, curr)

#STRINGOVI I KARAKTERI
        elif curr == '\'': #\escape za ' karakter
            lexeme = self.read_string()
            if len(lexeme)==1:
              token = Token(Class.CHAR, lexeme)
            else:
              token = Token(Class.STRING, lexeme)
#ово је компликација, паскал не препознаје "", већ само '', врвт мора да се хендлује

        elif curr == '+':
            token = Token(Class.PLUS, curr)
        elif curr == '*':
            token = Token(Class.STAR, curr)
       #elif curr == '%': MOD

        elif curr == '/':
            token = Token(Class.FWDSLASH, curr)



        elif curr == '!':
            curr = self.next_char()
            if curr == '=':
                token = Token(Class.NEQ, '!=')
            else:
                token = Token(Class.NOT, '!')
                self.pos -= 1


        elif curr == ':':
            curr = self.next_char()
            if curr == '=':
                token = Token(Class.ASSIGN, ':=')
            else:
                token = Token(Class.COLON,":")
                self.pos-=1
        elif curr == '=':
                token = Token(Class.EQ, '=')



        elif curr == '<':
            curr = self.next_char()
            if curr == '=':
                token = Token(Class.LTE, '<=')
            elif curr == '>':
                token = Token(Class.NEQ, '<>')
            else:
                token = Token(Class.LT, '<')
                self.pos -= 1
        elif curr == '>':
            curr = self.next_char()
            if curr == '=':
                token = Token(Class.GTE, '>=')
            else:
                token = Token(Class.GT, '>')
                self.pos -= 1
        elif curr == '(':
            token = Token(Class.LPAREN, curr)
        elif curr == ')':
            token = Token(Class.RPAREN, curr)
        elif curr == '[':
            token = Token(Class.LBRACKET, curr)
        elif curr == ']':
            token = Token(Class.RBRACKET, curr)
        elif curr == ';':
            token = Token(Class.SEMICOLON, curr)
        elif curr == ',':
            token = Token(Class.COMMA, curr)
        elif curr =='.':
          curr = self.next_char()
          if curr == '.':
            token = Token(Class.DOTDOT, '..')


        else:
            self.die(curr)
        #print(f"curr:{curr}")
        #print(token)
        return token

    def lex(self):
        self.tokens = []
        while True:
            curr = self.next_token()
            self.tokens.append(curr)


            if curr.class_ == Class.EOF:
                break
        return self.tokens

    def die(self, char):
        for t in self.tokens:
            print(t)
        #%tb
        raise SystemExit("Unexpected character:'{}'".format(char))
```

Klasa **Node** predstavlja baznu klasu za formiranje AST, a klase koje je nasleđuju odgovaraju svakoj ispravnoj semantičkoj strukturi.

```
class Node():
    pass


class Program(Node):
    def __init__(self, nodes):
        self.nodes = nodes

#декларације су облика
#i:integer
#i:array[a..b] of integer
#function fun1
#procedure proc1
class Decl(Node):
    def __init__(self, type_, id_):
        self.type_ = type_
        self.id_ = id_


class ArrayDecl(Node): #у цеу има величина низа, а овде од до
    def __init__(self, type_, id_, size, elems):
        self.type_ = type_
        self.id_ = id_
        self.size=size
        self.elems = elems   #за динамичку доделу елемената


class ArrayElem(Node):  #arrId[index] arr[3]
    def __init__(self, id_, index):
        self.id_ = id_
        self.index = index
        ind=index
        if type(index).__name__.lower()=='id':
          ind=index.value
        self.value=f"{id_.value}[{ind}]"


class Assign(Node):
    def __init__(self, id_, expr):   #:=
        self.id_ = id_
        self.expr = expr

#немамо else if
class If(Node):
    def __init__(self, cond, true, false):  #if cond  then true
        self.cond = cond                    #else false
        self.true = true
        self.false = false

#ПЕТЉЕ
class While(Node):  #уместо while имамо repeat until
    def __init__(self, cond, block):
        self.cond = cond
        self.block = block

class Repeat(Node):
    def __init__(self,cond,block):
        self.cond=cond
        self.block=block


class For(Node): #ово је другачије у паскалу него у цеу, степ је увек 1
    def __init__(self, init,cond, direction, block):
        self.parent = None
        self.init = init
        self.cond=cond
        self.direction = direction
        self.block = block


class FuncImpl(Node):
    def __init__(self, type_, id_, params, block, vars):
        self.type_ = type_ # void / int/char
        self.id_ = id_
        self.params = params
        self.block = block
        self.vars = vars

class Vars(Node):
    def __init__(self, declarations):
        self.declarations = declarations

class FuncCall(Node):
    def __init__(self, id_, args):
        self.parent=None
        self.id_ = id_
        self.args = args
        self.value=f"{id_.value}({args.value})"


class Block(Node):  #блок је чвор који садржи низ др чворова
    def __init__(self, nodes):
        self.parent = None
        self.nodes = nodes


class Params(Node):
    def __init__(self, params): #низ декларација параметра
        self.params = params


class Args(Node):
    def __init__(self, args):  #низ ID-ева променљивих неких
        self.args = args      #ТОДО паралелни низ прецизности
        val=""
        for i,a in enumerate(args):
          type_=type(a).__name__.lower
          if i>0:
            val+=','
          if type_ in ['str','int','char']:
            val+=a
          elif type_ in ['id','string','int','integer','char']:
            val+=a.value

        self.value=val
        self.parent = None


class Elems(Node):   #елементи су arr[1..3]=(1,2,3)
    def __init__(self, elems): #зарезима су одвојени expression-и
        self.elems = elems


class Break(Node):
    pass


class Continue(Node):
    pass


class Return(Node):
    def __init__(self, expr):
        self.expr = expr #произвољно


class Type(Node): #integer,char,string
    def __init__(self, value):
        self.value = value

class Argument(Node):
    def __init__(self, arg, round):
        self.arg=arg
        self.round=round
        #self.value=arg.value
        self.parent = None

class Round(Node):
    def __init__(self, first, second):
        self.first=first
        self.second=second

class Int(Node):      #литерал број
    def __init__(self, value):
        self.value = value

class Direction(Node):
    def __init__(self,var, value):
        self.var=var
        self.value = value
        self.increment=None


class Char(Node):         #литерал чар
    def __init__(self, value):
        self.value = value


class String(Node):       #литерал стринг
    def __init__(self, value):
        self.value = value


class Id(Node):
    def __init__(self, value):
        self.id_=value
        self.value = value



#од операција се праве изрази
class BinOp(Node):
    def __init__(self, symbol, first, second):
        self.symbol = symbol
        self.first = first
        self.second = second
        s=symbol
        if s == 'mod':
            s='%'
        elif s=='div':
            s='/'
        self.id_=Id(f"{first.value} {s} {second.value}")
        self.value=self


class UnOp(Node):
    def __init__(self, symbol, first):
        self.symbol = symbol
        self.first = first

#ТОДО у паскалу има кључна реч edit
```

```
class Parser:

    def die_type(self, expected, found):
        self.die("Expected: {}, Found: {}".format(expected, found))

    def __init__(self, tokens):
        self.tokens = tokens
        self.curr = tokens.pop(0) #instance attribute
        self.prev = None          #instance attribute

    def restorable(call):#не мењамо, служи за враћање прошлог стања листе токена
        @wraps(call)
        def wrapper(self, *args, **kwargs):
            state = pickle.dumps(self.__dict__)
            result = call(self, *args, **kwargs)
            self.__dict__ = pickle.loads(state)
            return result
        return wrapper

    def eat(self, class_):  #не мењамо
        if self.curr.class_ == class_:
            self.prev = self.curr
            self.curr = self.tokens.pop(0)
        else:
            self.die_type(class_.name, self.curr.class_.name)
        print("pojedeno:"+str(self.prev.lexeme))

#void fun()
    def program(self):
        nodes = []
        while self.curr.class_ != Class.EOF:
            if self.curr.class_ == Class.TYPE:   #свака целина почиње с type
                nodes.extend(self.decl())        #у цеу, али у паскалу,
                                                 #може са VAR,TYPE, BEGIN-END blokovima
            elif self.curr.class_==Class.VAR:
                self.eat(Class.VAR)
                while self.curr.class_!=Class.BEGIN:
                  nodes.extend(self.decl())

            elif self.curr.class_==Class.BEGIN:
              self.eat(Class.BEGIN)
              block=self.block()
              self.eat(Class.FEND)
              nodes.append(block)
            #   nodes.extend(block.nodes)

            else:
                self.die_deriv(self.program.__name__)
        return Program(nodes)




                        #за читање свега што почиње с ИД
                        #у цеу, позиф ф-је, додела, пок. на еле низа
                        #fin(3+1); i; i=10; i[2+3];
    def id_(self):
        #is_array_elem = self.prev.class_ != Class.TYPE
        id_ = Id(self.curr.lexeme)
        self.eat(Class.ID)
        if self.curr.class_ == Class.LPAREN and self.is_func_call(): #poziv fje
            self.eat(Class.LPAREN) #x
            print(f"id {id_.value} is func call")
            args = self.args()
            print("pojedeni argumenti ",args)
            self.eat(Class.RPAREN)
            #if id_ in ['ord','chr']:
            #  return args
            return FuncCall(id_, args)
        elif self.curr.class_ == Class.LBRACKET:# and is_array_elem: #index arr
            self.eat(Class.LBRACKET) #x[expr]
            index = self.expr()
            self.eat(Class.RBRACKET)
            #print(f"id is {id_.value}[{index.value}]")
            id_ = ArrayElem(id_, index)
            print('id_=',id_.value)
        if self.curr.class_ == Class.ASSIGN: #x:=expr; x[expr]:=expr
            self.eat(Class.ASSIGN)
            #print("id:=,  ",id_.value,":=")
            expr = self.logic()
            return Assign(id_, expr)
        else:
            return id_

    def readVars(self):
        self.eat(Class.VAR)
        declarations = []
        while self.curr.class_ != Class.BEGIN:
            d=self.decl()
            declarations.extend(d)
        return Vars(declarations)


#декларације moze biti 1.deklaracija +(dodela)? 2.f-ja
#i,j,k:integer (=expr)?;               ID,TYPE,EXPR
#i:array[a..b] of integer ()=elemx)?;  ID,ARRAY,[,1ST,dotdot, LAST,], OF, TYPE
#function fun1(a,b:TYPE)  TYPE,ID,(,ID,ID,COLON,TYPE,)
#procedure proc1(a,b:TYPE):rtype  TYPE,ID,(,ID,ID,COLON,TYPE,),TYPE
                  #дакле, прво иде ИЛИ тип ИЛИ ид
                  #
    def decl(self):   #svi oni koji pocinju sa TYPE su ovde obuhvaceni
                     #procedure, function         #x,y (:integer;s)
        lexeme=self.curr.lexeme.lower()
        #PROCEDURE/FUNCTION
        if lexeme=="procedure" or lexeme=="function":
          type_=self.type_()
          id_=self.id_()
          if self.curr.class_==Class.LPAREN:
            self.eat(Class.LPAREN)
            params=self.params()      #TODO
            self.eat(Class.RPAREN)
                                            #разкујзхени функц и проц
            if self.curr.class_==Class.SEMICOLON:
              self.eat(Class.SEMICOLON)
            elif self.curr.class_==Class.COLON:
              self.eat(Class.COLON)
              type_=self.type_()         #ako je procedura, type ostje procedure
              self.eat(Class.SEMICOLON)#inace type postaje ono sto vraca

            if self.curr.class_==Class.VAR:
              vars=self.readVars()
            else:
              vars=None

            self.eat(Class.BEGIN)
            block = self.block()
            self.eat(Class.END)
            self.eat(Class.SEMICOLON)

            d = []
            d.append(FuncImpl(type_,id_,params,block,vars))
            return d

        #INTEGER AND STRING (MULTI), ARRAY
        elif self.curr.class_==Class.ID:
          id_=self.id_()
          IDs=[]
          IDs.append(id_)
          #print("curr.class_ je ID, i leksema je:"+self.prev.lexeme)

          if (self.curr.class_==Class.COMMA or self.curr.class_==Class.COLON) and tokens[1].lexeme!="array":
            if self.curr.class_==Class.COMMA: #ako je obicna dekl
              self.eat(Class.COMMA)
              while self.curr.class_ != Class.COLON: #pravimo spisak var, do :
                if self.curr.class_==Class.ID:
                  IDs.append(self.id_())     #a,b,c
                if self.curr.class_==Class.COMMA:
                  self.eat(Class.COMMA)
              self.eat(Class.COLON)               #:
            else:self.eat(Class.COLON) #a:

            #AKO NIJE ARRAY
            if self.curr.class_==Class.TYPE :
              type_=self.type_()                 #integer/string[exp]

            #ARRAY DECL
            elif self.curr.class_==Class.ARRAY:
              self.eat(Class.ARRAY)
              self.eat(Class.LBRACKET)
              first=self.expr()
              self.eat(Class.DOTDOT)
              last=self.expr()
              self.eat(Class.RBRACKET)
              self.eat(Class.OF)
              type_=self.type_()            #arr1:array[FIRST..LAST] OF TYPE
              elems=None
              size=Int(last.value)# - first.value)

              if self.curr.class_==Class.EQ:
                self.eat(Class.EQ)
                self.eat(Class.LPAREN)
                elems=self.elems()          #=(1,2,3)?
                self.eat(Class.RPAREN)

              self.eat(Class.SEMICOLON)     #;

              d=[]
              for id in IDs:
                d.append(ArrayDecl(type_,id,size,elems))
              return d

            else:
              self.die_type("TYPE or ARRAY",self.curr.lexeme)

            array=False
            if type_.value.lower()=="string":         #string[exp]
              if self.curr.class_==Class.LBRACKET:
                self.eat(Class.LBRACKET)
                size_=self.expr()
                self.eat(Class.RBRACKET)
                array=True

            self.eat(Class.SEMICOLON)           #;

            declarations=[]
            if type_.value.lower()=="string":
              for i in IDs:
                if array:
                  declarations.append(ArrayDecl(type_,i,size_,None))
                else:
                  declarations.append(Decl(type_,i))
            else:
              for i in IDs:
                declarations.append(Decl(type_,i)) #била обрнута места тајп и ИД
            return declarations





#у цеу ду декл облика
#int x; V int x=1;
#int x[3] V int x[3]={1,2,3}
#void fun1(Type ID){}
#int fun2(){}

#деклрарације у паскалу се налазе само у вар делу

#if(x==5){} u ceu

#if x=5 then begin end        u paskalu
#IF LOGIC BLOCK
    def if_(self):
        self.eat(Class.IF)
        cond = self.logic()
        print('cond:',cond.id_)
        self.eat(Class.THEN)
        self.eat(Class.BEGIN)
        true = self.block()
        self.eat(Class.END)
        false = None
        if self.curr.class_ == Class.ELSE:
            self.eat(Class.ELSE)
            self.eat(Class.BEGIN)
            false = self.block()
            self.eat(Class.END)

        self.eat(Class.SEMICOLON)
        return If(cond, true, false)

    def while_(self):
        self.eat(Class.WHILE)
        cond = self.logic()
        self.eat(Class.DO)
        self.eat(Class.BEGIN)
        block = self.block()
        self.eat(Class.END)
        self.eat(Class.SEMICOLON)
        return While(cond, block)

#todo repeat until
    def repeat(self):
        self.eat(Class.REPEAT)
        block=self.block()
        self.eat(Class.UNTIL)
        cond=self.logic()
        self.eat(Class.SEMICOLON)
        return Repeat(cond,block)

    def for_(self):
        self.eat(Class.FOR)
        init = self.id_()

        operation=""
        symbol=None
        if self.curr.class_==Class.TO:
          direction=Direction(init.id_.value,1)
          operation="+"
          symbol="<="
          self.eat(Class.TO)
        elif self.curr.class_==Class.DOWNTO:       #TODO
          direction=Direction(init.id_.value,-1)
          operation="-"
          symbol=">="
          self.eat(Class.DOWNTO)

        second=self.expr()
        print("second:",second.value)
        cond=BinOp(symbol,init.id_,second)

        self.eat(Class.DO)
        self.eat(Class.BEGIN)
        block = self.block()
        self.eat(Class.END)
        self.eat(Class.SEMICOLON)


        operation=BinOp(operation,init.id_,Int(1)) #i +- 1
        increment=Assign(init.id_,operation) #i:=i +- 1
        direction.increment=increment


        return For(init, cond, direction, block)

#садржи све остале
#тодо убацити repeat until
    def block(self):
        nodes = []
        while self.curr.class_ not in [Class.END,Class.FEND,Class.EOF,Class.UNTIL]:# or self.curr.class_!=Class.FEND: TODO
            if self.curr.class_ == Class.IF:
                nodes.append(self.if_())
            elif self.curr.class_ == Class.WHILE:
                nodes.append(self.while_())
            elif self.curr.class_ == Class.FOR:
                nodes.append(self.for_())
            elif self.curr.class_ == Class.REPEAT:
                nodes.append(self.repeat())
            elif self.curr.class_ == Class.BREAK:
                nodes.append(self.break_())
            elif self.curr.class_ == Class.CONTINUE:
                nodes.append(self.continue_())
            elif self.curr.class_ == Class.EXIT:
                nodes.append(self.exit_())
            elif self.curr.class_ == Class.TYPE:
                nodes.append(self.decl())
            elif self.curr.class_ == Class.ID:
                nodes.append(self.id_())
                self.eat(Class.SEMICOLON)

            else:
                print("curr token class:"+str(self.curr.class_))
                self.die_deriv(self.block.__name__)
        return Block(nodes)





#писемо код који поштује регекс x,y,z:integer нпр
#у овом коду нпр main("int x, int y"){}
    def params(self):
        params = []
        while self.curr.class_ != Class.RPAREN:
          vars=[]
          while self.curr.class_!=Class.COLON :
            if len(vars) > 0:
                self.eat(Class.COMMA)
            vars.append(self.id_())
          self.eat(Class.COLON)
          type_=self.type_()

          for v in vars:
            params.append(Decl(type_,v))

        return Params(params)

    def args(self):
        args = []
        Uzagradi=False
        while self.curr.class_ != Class.RPAREN:
            if len(args) > 0:
                self.eat(Class.COMMA)

            if self.curr.class_==Class.LPAREN:
                self.eat(Class.LPAREN)
                Uzagradi=True

            arg = self.expr()

            if Uzagradi:
                self.eat(Class.RPAREN)
                Uzagradi=False

            if self.curr.class_== Class.COLON:
                self.eat(Class.COLON)
                first=self.expr()
                self.eat(Class.COLON)
                second=self.expr()
                round=Round(first,second)
            else:
                round=None

            print("apendovan argument: ",arg,round)
            args.append(Argument(arg,round))

        return Args(args)

#arr1[1..3]=(1,2,3) elems=1,2,3
    def elems(self):
        elems = []
        while self.curr.class_ != Class.RPAREN:
            if len(elems) > 0:
                self.eat(Class.COMMA)
            elems.append(self.expr())
        return Elems(elems)

    def exit_(self):
        self.eat(Class.EXIT)
        expr=None
        if self.curr.class_==Class.LPAREN:self.eat(Class.LPAREN)
        expr = self.expr()
        if self.curr.class_==Class.RPAREN:self.eat(Class.RPAREN)
        self.eat(Class.SEMICOLON)
        return Return(expr)

    def break_(self):
        self.eat(Class.BREAK)
        self.eat(Class.SEMICOLON)
        return Break()

    def continue_(self):
        self.eat(Class.CONTINUE)
        self.eat(Class.SEMICOLON)
        return Continue()

    def type_(self):
        type_ = Type(self.curr.lexeme)
        self.eat(Class.TYPE)
        return type_

    def factor(self):
      #читање литерала
        if self.curr.class_ == Class.INTEGER:
            value = Int(self.curr.lexeme)
            self.eat(Class.INTEGER)
            return value
        elif self.curr.class_ == Class.CHAR:
            value = Char(self.curr.lexeme)
            self.eat(Class.CHAR)
            return value
        elif self.curr.class_ == Class.STRING:
            value = String(self.curr.lexeme)
            self.eat(Class.STRING)
            return value
        elif self.curr.class_ == Class.TRUE:
            value =Int(1)
            self.eat(Class.TRUE)
            return value
        elif self.curr.class_ == Class.FALSE:
            value = Int(0)
            self.eat(Class.FALSE)
            return value

      #читање ИД-а
        elif self.curr.class_ == Class.ID:
            return self.id_()
        elif self.curr.class_ in [Class.MINUS, Class.NOT, Class.ADDRESS]:
            op = self.curr
            self.eat(self.curr.class_)
            first = None
            if self.curr.class_ == Class.LPAREN:
                self.eat(Class.LPAREN)
                first = self.logic()
                self.eat(Class.RPAREN)
            else:
                first = self.factor()
            return UnOp(op.lexeme, first)
        elif self.curr.class_ == Class.LPAREN:
            self.eat(Class.LPAREN)
            first = self.logic()
            self.eat(Class.RPAREN)
            return first
        elif self.curr.class_ == Class.SEMICOLON:
            return None
        else:
            self.die_deriv(self.factor.__name__)

    def term(self):
        first = self.factor()
        while self.curr.class_ in [Class.STAR, Class.FWDSLASH, Class.DIV, Class.MOD]:
            if self.curr.class_ == Class.STAR:
                op = self.curr.lexeme
                self.eat(Class.STAR)
                second = self.factor()
                first = BinOp(op, first, second)
            elif self.curr.class_ == Class.FWDSLASH:
                op = self.curr.lexeme
                self.eat(Class.FWDSLASH)
                second = self.factor()
                first = BinOp(op, first, second)
            elif self.curr.class_ == Class.DIV:
                op = self.curr.lexeme
                self.eat(Class.DIV)
                second = self.factor()
                first = BinOp(op, first, second)
            elif self.curr.class_ == Class.MOD:
                op = self.curr.lexeme
                self.eat(Class.MOD)
                second = self.factor()
                first = BinOp(op, first, second)
        return first

    def expr(self):
        first = self.term()
        while self.curr.class_ in [Class.PLUS, Class.MINUS]:
            if self.curr.class_ == Class.PLUS:
                op = self.curr.lexeme
                self.eat(Class.PLUS)
                second = self.term()
                first = BinOp(op, first, second)
            elif self.curr.class_ == Class.MINUS:
                op = self.curr.lexeme
                self.eat(Class.MINUS)
                second = self.term()
                first = BinOp(op, first, second)
        return first

    def compare(self):
        first = self.expr()
        if self.curr.class_ == Class.EQ:
            op = self.curr.lexeme
            self.eat(Class.EQ)
            second = self.expr()
            return BinOp(op, first, second)
        elif self.curr.class_ == Class.NEQ:
            op = self.curr.lexeme
            self.eat(Class.NEQ)
            second = self.expr()
            return BinOp(op, first, second)
        elif self.curr.class_ == Class.LT:
            op = self.curr.lexeme
            self.eat(Class.LT)
            second = self.expr()
            return BinOp(op, first, second)
        elif self.curr.class_ == Class.GT:
            op = self.curr.lexeme
            self.eat(Class.GT)
            second = self.expr()
            return BinOp(op, first, second)
        elif self.curr.class_ == Class.LTE:
            op = self.curr.lexeme
            self.eat(Class.LTE)
            second = self.expr()
            return BinOp(op, first, second)
        elif self.curr.class_ == Class.GTE:
            op = self.curr.lexeme
            self.eat(Class.GTE)
            second = self.expr()
            return BinOp(op, first, second)
        else:
            return first

    def logic_term(self):
        first = self.compare()
        while self.curr.class_ == Class.AND:
            op = self.curr.lexeme
            self.eat(Class.AND)
            second = self.compare()
            first = BinOp(op, first, second)
        return first

    def logic(self):
        first = self.logic_term()
        while self.curr.class_ == Class.OR:
            op = self.curr.lexeme
            self.eat(Class.OR)
            second = self.logic_term()
            first = BinOp(op, first, second)
        return first

    @restorable
    def is_func_call(self):
        try:
            self.eat(Class.LPAREN)
            self.args()
            self.eat(Class.RPAREN)
            return self.curr.class_ in [Class.SEMICOLON,Class.COMMA,Class.THEN,Class.LTE,Class.GTE,Class.NEQ,Class.EQ,Class.GT,Class.LT,Class.DIV,Class.MOD,Class.MINUS,Class.PLUS,Class.RPAREN,Class.LPAREN] #акое ; ондае позив
        except:                                 #исто НЕ важи за паскал
            return False

#служи да разликујемо позив функције од импл функције
#main(); овде restorable враћа true
#VS main(){} овде враћа false
                            #ОВО ЈЕ ПРИМЕР ГЛЕДЊА У НАПРЕД ЗА БЛОКОВЕ
    def parse(self):
        return self.program()

    def die(self, text):
        print("current lexeme:"+str(self.curr.lexeme))
        raise SystemExit(text)

    def die_deriv(self, fun):
        self.die("Derivation error: {}".format(fun))



```

```
class Visitor():
    def visit(self, parent, node):
        method = 'visit_' + type(node).__name__
        visitor = getattr(self, method, self.die)
        #print(method)
        return visitor(parent, node)

    def die(self, parent, node):
        method = 'visit_' + type(node).__name__
        raise SystemExit("Missing method: {}".format(method))
```

```
class Generator(Visitor):
    def __init__(self, ast):
        self.ast = ast
        self.c = ""
        self.level = 0

    def append(self, text):
        self.c += str(text)

    def newline(self):
        self.append('\n\r')

    def indent(self):
        for i in range(self.level):
            self.append('\t')

    def visit_Program(self, parent, node):
      main=False
      funkcije=[]
      for i,n in enumerate(node.nodes):
        if type(n).__name__ == 'FuncImpl':
          if i in funkcije:
            self.visit(node, n)
            funkcije.append(i)
        elif not main:
          main=True
          self.append("int main(){")
          self.newline()
        self.visit(node,n)
        if i==len(node.nodes)-1:
         self.append('\nreturn 0;')
         self.append('\n}')
         self.newline()

    def visit_Decl(self, parent, node):
        self.visit(node, node.type_)
        #self.append(f"note.type_={node.type_.value}")
        self.append(" ")
        self.visit(node, node.id_)
        self.append(";\n")

    def visit_ArrayDecl(self, parent, node):
        self.visit(node, node.type_)
        self.append(" ")
        self.visit(node, node.id_)
        if node.elems is not None:
            self.append('[] = {')
            self.visit(node, node.elems)    #arr2[]={1,2,3};
            self.append('};\n')
        elif node.size is not None:
            self.append('[')
            self.visit(node, node.size)     #arr3[3]
            self.append('];\n')

    def visit_ArrayElem(self, parent, node):
        self.visit(node, node.id_)
        self.append('[')
        self.visit(node, node.index)
        self.append(']')

    def visit_Assign(self, parent, node):
        self.visit(node, node.id_)
        self.append(' = ')
        if type(node.expr).__name__ == 'String':
          self.append('"')
          self.visit(node, node.expr)
          self.append('"')
        else:
          self.visit(node, node.expr)
        if type(node.expr).__name__ != 'FuncCall':
          self.append(';')
        if type(parent).__name__ != "For":
          self.append("\n")

    def visit_If(self, parent, node):
        self.append('if (')
        self.visit(node, node.cond)
        self.append('){')
        self.newline()
        self.visit(node, node.true)
        self.append('}')
        if node.false is not None:
            self.indent()
            self.append('else {')
            self.newline()
            self.visit(node, node.false)
            self.append("}")

    def visit_While(self, parent, node):
        self.append('while (')
        self.visit(node, node.cond)
        self.append('){')
        self.newline()
        self.visit(node, node.block)
        self.append('}')

    def visit_Repeat(self,parent,node):
        self.append('do{\n')
        self.visit(node, node.block)
        self.append('} \nwhile(!')
        self.visit(node,node.cond)
        self.append(');\n')

    def visit_For(self, parent, node):
        self.newline()
        self.indent()
        self.append('for (')
        self.visit(node, node.init)
        self.append(" ")
        self.visit(node, node.cond)
        self.append("; ")
        self.visit(node, node.direction)
        self.append('){')
        self.newline()
        self.visit(node, node.block)
        self.level += 1
        self.indent()
        self.level -= 1
        self.append("}\n")

    def visit_Vars(self, parent, node):
        self.indent()
        for v in node.declarations:
          self.visit(node.declarations, v)

    def visit_FuncImpl(self, parent, node):
        if node.type_.value.lower()=='procedure':self.append('void ')
        else:
          self.visit(node, node.type_)
        self.append(f" {node.id_.value} ")
        self.append('(')
        self.visit(node, node.params)
        self.append('){\n')
        if node.vars is not None:
          self.visit(node, node.vars)
        self.newline()

        #self.append(f"\n\t\t\t node.type_.value.lower()={node.type_.value.lower()} \n")

        if node.type_.value.lower()!='procedure':
          self.append(f"\t{node.type_.value} {node.id_.value};")
        self.newline()
        self.visit(node, node.block)
        if node.type_.value.lower()!='procedure':
          self.append(f"\treturn {node.id_.value};")
        self.append("\n}\n")

    def calcType(self,first, second):
      print(f"calcType, first:{first} second:{second}")
      if first.lower() in ['float','real'] or second.lower() in ['float', 'real']:
        return 'float'
      elif first.lower() =='int' or second.lower()=='int':
        return 'int'
      else:
        return 'char'

    def findType(self,id_,node):
      while type(node).__name__.lower() not in ['funcimpl','program']:
        node=node.parent

      declarations=[]
      if type(node).__name__.lower() == 'funcimpl':
        declarations.extend(node.vars.declarations)
        if node.params is not None:
          declarations.extend(node.params)
      else:
        for n in node.nodes:
          if type(n).__name__.lower() in ['decl','arraydecl']:
            declarations.append(n)
          elif type(n).__name__.lower() in ['funcimpl']:
            declarations.append(Decl(n.type_,n.id_))

      for d in declarations:
        #print(f"id_={id_}, d.id.value={d.id_.value}", end=' ')
        if id_==d.id_.value:
          return d.type_.value

      return 'undeclaired'

    def getType(self,parent,node):
      if type(node).__name__ in ['Int','Char','String']:
        return type(node).__name__

      elif type(node).__name__.lower() in ['id','arrayelem']:
          id_=node.value
          if type(node).__name__.lower() =='arrayelem':
            id_=node.id_.value

          type_=self.findType(id_,parent)
          if type_ != 'undeclaired':
            return type_
          else:
            print(f"variable {id_} out of scope")

      elif type(node).__name__.lower()=='funccall':
          scope=node
          chrord=False
          while type(scope).__name__.lower()!='program':

            print(f"TYPE SCOPE NAME LOWER: {type(scope).__name__.lower()}")
            if type(node).__name__.lower()=='funccall':
              print("func:",node.id_.value)
              func=True
              if node.id_.value == 'chr':
                type_='char'
              elif node.id_.value=='ord':
                print(f"is ord", end=' ')
                type_='int'
              else:
                type_=self.findType(node.id_.value,scope)
                print(f"f-ja je {node.id_.value}")

            scope=scope.parent
          #type_ = scope.get(node.id_).type_
          if func:
            return type_
          type_ = self.findType(node.id_, scope)
          return type_

      elif type(node).__name__=='BinOp':
          if node.symbol in ['==','<=','>=','<>','<','>']:
            raise systemexit(f"logical expression found, but expected arithmetic")
          else:
            first_type = self.getType(parent,node.first)
            second_type = self.getType(parent,node.second)
            print(f"type(first).__name__={type(node.first).__name__} \ntype(second).__name__={type(node.second).__name__}")
            return self.calcType(first_type,second_type)

      else:
        raise SystemExit(f"cannot print type {type(node).__name__}")

    def visit_FuncCall(self, parent, node):
        node.parent=parent
        print(f"GENERATOR: parent of {type(node).__name__} is {type(node.parent).__name__}")
        func = node.id_.value
        args = node.args.args
        if func == 'writeln' or func == 'write':

          self.append('printf("')
          dva=[1,2]
          hasArgs=False
          for j,x in enumerate(dva):
            for i,a in enumerate(args):
              hasArgs=True
              #print(f"args.arg[i]: {a.arg} \ntype(args.arg).__name__ :{type(a.arg).__name__}")

              if j==0:
                type_=self.getType(parent,a.arg)
                if type(a.arg).__name__.lower() in ['string','int','char']:
                  self.append(f"{a.arg.value}")

                elif type_.lower() in ['int','integer']:
                  self.append('%d')
                elif type_.lower() == 'char':
                  self.append('%c')
                #elif type_.lower() == 'string':
                  #self.append('%s')

                elif type_.lower() in ['real','float']:
                  if round is not None:
                    self.append(f"%.{a.round.second.value}f")
                  else:
                    self.append("%f")
                else:
                  print(f"invalid type:{type_}")


              if j==1:
                if i==0:
                  if func == 'writeln':
                    self.append("\\n")
                  self.append('"')
                if type(a.arg).__name__.lower() not in ['str','string','char']:
                  if type(a.arg.id_).__name__.lower() not in ['str','string','char']:
                    #variables+=f",{a.arg.id_.value}"
                    self.append(",")
                    self.visit(a,a.arg)
                  #self.append(f"visit_{type(node).__name__}")
                  else:
                    self.append(f",{a.arg.id_}")

              #else:
                #self.append(a.arg.value)

          if func == 'writeln' and not hasArgs:
            self.append("\\n")
            self.append('"')
          self.append(')')

        elif func in ['read','readln']:
            self.append('scanf("')
            declarations=[]

            while type(parent).__name__.lower() not in ['funcimpl','program']:
                parent=parent.parent
            if type(parent).__name__.lower()=='funcimpl':     #ДОХВАТАЊЕ ДЕКЛАРАЦИЈА ИЗ ЛОКАЛНОГ ОПСЕГА
                declarations=parent.vars.declarations
            elif type(parent).__name__.lower()=='program':
                for n in parent.nodes:
                  if type(n).__name__.lower()=='decl':
                    declarations.append(n)
                  elif type(n).__name__.lower()=='arraydecl':
                    declarations.append(n)
            else:
                raise SystemExit("incorrect syntax tree")

            IDs=[]
            arguments=[]
            for d in declarations:
              #print("for d in declarations, d.id_.value=",d.id_.value)
              IDs.append(d.id_.value)
              print("declaired ids:",d.id_.value)
            for a in node.args.args:
              #print("for a in node.args.args, a.arg.value=",a.arg.value)     #ДОХВАТАЊЕ ИД-ЕВА И ПРОВЕРА ДАЛ СУ АРГУМЕНТИ У ЊИМА
              arguments.append(a.arg.value)
              if type(a.arg).__name__.lower()=='arrayelem':
                arg=a.arg.id_.value
              else:
                arg=a.arg.value
              if arg not in IDs:
                #print(f"argument:{a.arg.value}")
                raise SystemExit(f"variable {arg} not declaired or out of scope")

            ampercent=""
            for a in node.args.args:
              for d in declarations:
                if type(a.arg).__name__.lower()=='arrayelem':
                  arg=a.arg.id_.value
                else:
                  arg=a.arg.value
                if d.id_.value == arg:
                  type_=d.type_.value
                  if type_=='real':
                    self.append('%f')
                  elif type_=='float':
                    self.append('%f')
                  elif type_ in ['integer','int']:
                    self.append('%d')
                  elif type_=='string':
                    self.append('%s')
                  elif type_=='char':
                    self.append('%c')
                  if type(a.arg).__name__.lower()=='arrayelem':
                    arg=a.arg.value
                  ampercent+=f", &{arg}"
            self.append('"')
            if ampercent != "":
              self.append(ampercent)

            self.append(')')
        elif func == 'length':
            self.append('strlen(')
            self.visit(node, node.args)
            self.append(')')
        elif func == 'concat':
            self.append('strcat(')
            self.visit(node.args, args[0])
            self.append(',')
            self.visit(node.args, args[1])
            self.append(')')
            self.indent()
        elif func.lower() in ['ord','chr']:
            arg=node.args.args[0].arg
            self.visit(node, node.args)
        else:
            self.append(func)
            self.append('(')
            self.visit(node, node.args)
            self.append(')')
        if type(parent).__name__.lower() not in ['if','for','binop','funccall','argument','return']:
            #self.append(f"type(parent).__name__.lower()={type(parent).__name__.lower()}")
            self.append(';\n')

    def visit_Block(self, parent, node):
        self.level += 1
        for n in node.nodes:
            self.indent()
            self.visit(node, n)
            self.newline()
        self.level -= 1

    def visit_Params(self, parent, node):
        for i, p in enumerate(node.params):
            if i > 0:
                self.append(', ')
            self.visit(p, p.type_)
            self.append(" ")
            self.visit(p, p.id_)

    def visit_Args(self, parent, node):
        for i, a in enumerate(node.args):
            if i > 0:
                self.append(', ')
            node.parent = parent
            self.visit(node, a)
            #self.append(f"visit_{type(a).__name__}")

    def visit_Argument(self,parent,node):
        node.parent = parent
        self.visit(node,node.arg)
        #self.append(f"type(argument.arg).__name__ = {type(node.arg).__name__.lower()}")

    def visit_Elems(self, parent, node):
        for i, e in enumerate(node.elems):
            if i > 0:
                self.append(', ')
            self.visit(node, e)

    def visit_Break(self, parent, node):
        self.append('break;')

    def visit_Continue(self, parent, node):
        self.append('continue;')

    def visit_Return(self, parent, node):
        self.append('return')
        if node.expr is not None:
          self.append(' ')
          self.visit(node, node.expr)
          #self.append(f"ne teet, type:{type(node.expr).__name__}")
        self.append(";")

    def visit_Type(self, parent, node):
        if node.value in ['integer','boolean']:
          node.value='int'
        elif node.value=='real':
          node.value='float'
        self.append(node.value)

    def visit_Int(self, parent, node):
        self.append(node.value)

    def visit_Direction(self, parent, node):
        self.append(node.var)
        if node.value==1:
          self.append('++')
        else:
          self.append('--')

    def visit_Char(self, parent, node):
        self.append("'")
        self.append(node.value)
        self.append("'")

    def visit_String(self, parent, node):
        self.append(node.value)

    def visit_Id(self, parent, node):
        self.append(node.value)

    def visit_BinOp(self, parent, node):
        self.append('(')
        self.visit(node, node.first)
        if node.symbol == 'AND':
            self.append(' && ')
        elif node.symbol == 'OR':
            self.append(' || ')
        elif node.symbol.lower() == 'mod':
            self.append(' % ')
        elif node.symbol.lower() == 'div':
            self.append(' / ')
        elif node.symbol == '=':
            self.append(" == ")
        elif node.symbol == '<>':
            self.append(' != ')
        else:
            self.append(node.symbol)
        self.visit(node, node.second)
        self.append(')')

    def visit_UnOp(self, parent, node):
        if node.symbol == '!':
            self.append('!')
        elif node.symbol != '&':
            self.append(node.symbol)
        self.visit(node, node.first)

    def put(self,node, id_, type_, scope):
        node.symbols[id_] = Symbol(id_, type_, scope)

    def get(self,node, id_):
        return node.symbols[id_]

    def contains(self,node, id_):
      print("id_",id_," compared to: ")
      for symbol in node.symbols:
        print("symbol")
        return id_ in node.symbols

    def generate(self, path):
        self.visit(None, self.ast)
        self.c = re.sub('\n\s*\n', '\n', self.c)
        with open(path, 'w') as source:
            source.write(self.c)
        return path
```

```
class Symbol:
    def __init__(self, id_, type_, scope):
        self.id_ = id_
        self.type_ = type_
        self.scope = scope

    def __str__(self):
        return "<{} {} {}>".format(self.id_, self.type_, self.scope)

    def copy(self):
        return Symbol(self.id_, self.type_, self.scope)
```

```
class Symbols:
    def __init__(self):
        self.symbols = {}

    def put(self, id_, type_, scope):
        self.symbols[id_] = Symbol(id_, type_, scope)

    def get(self, id_):
        return self.symbols[id_]

    def contains(self, id_):
        return id_ in self.symbols

    def remove(self, id_):
        del self.symbols[id_]

    def __len__(self):
        return len(self.symbols)

    def __str__(self):
        out = ""
        for _, value in self.symbols.items():
            if len(out) > 0:
                out += "\n"
            out += str(value)
        return out

    def __iter__(self):
        return iter(self.symbols.values())

    def __next__(self):
        return next(self.symbols.values())
```

```
class Runner(Visitor):
    def __init__(self, ast):
        self.ast = ast
        self.global_ = {}
        self.local = {}
        self.scope = {}
        self.call_stack = []
        self.search_new_call = True
        self.return_ = False


        self.linija = None
        self.linijaIndex = 0


    # local: {
    #     139667859551120: [
    #         {
    #             'a': Symbol,
    #             'b': Symbol
    #          }
    #     ]
    #  }

    #  scope: {'program': []}
    #  call_stack: ['program']

    def get_symbol(self, node):
        recursion = self.is_recursion()
        ref_call = -2 if not self.search_new_call else -1
        ref_scope = -2 if recursion and not self.search_new_call else -1
        id_ = node.value
        if len(self.call_stack) > 0:
            fun = self.call_stack[ref_call]
            for scope in reversed(self.scope[fun]):
                if scope in self.local:
                    curr_scope = self.local[scope][ref_scope]
                    if id_ in curr_scope:
                        return curr_scope[id_]
        return self.global_[id_]

    def init_scope(self, node):
        fun = self.call_stack[-1]
        if fun not in self.scope:
            self.scope[fun] = []
        scope = id(node)
        if scope not in self.local:
            self.local[scope] = []
        self.local[scope].append({})
        for s in node.symbols:
            self.local[scope][-1][s.id_] = s.copy()

    def clear_scope(self, node):
        scope = id(node)
        self.local[scope].pop()

    def is_recursion(self):
        if len(self.call_stack) > 0:
            curr_call = self.call_stack[-1]
            prev_calls = self.call_stack[:-1]
            for call in reversed(prev_calls):
                if call == curr_call:
                    return True
        return False

    def visit_Program(self, parent, node):
        print("PROGRAM SYMBOLS:", node.symbols)
        for s in node.symbols:
            self.global_[s.id_] = s.copy()
        s = Symbol('program', '', id(node))
        self.global_['program'] = s
        # self.call_stack.append('program')

        for n in node.nodes:
            if isinstance(n, Block):
                self.call_stack.append('program')
                node.block = n
                s.block = n
                self.init_scope(n)
            print("PROGRAM NODE:", type(n).__name__)

            # self.visit(node, n)
            # if isinstance(n, Block):
            #     self.clear_scope(node)
            self.visit(node, n)
        self.clear_scope(node.block)
        self.call_stack.pop()

    def visit_Decl(self, parent, node):
        id_ = self.get_symbol(node.id_)
        id_.value = None

    def visit_ArrayDecl(self, parent, node):
        id_ = self.get_symbol(node.id_)
        id_.symbols = node.symbols
        size, elems = node.size, node.elems
        # if elems is not None:
        #     self.visit(node, elems)
        for i in range(size.value):
            id_.symbols.put(i, id_.type_, None)
            id_.symbols.get(i).value = None

    def visit_ArrayElem(self, parent, node):
        id_ = self.get_symbol(node.id_)
        index = self.visit(node, node.index)
        if isinstance(index,int):
            return id_.symbols.get(index)
        return id_.symbols.get(index.value)

    def visit_Assign(self, parent, node):
        id_ = self.visit(node, node.id_)
        value = self.visit(node, node.expr)
        if isinstance(value, Symbol):
            value = value.value
        id_.value = value

    def visit_If(self, parent, node):
        cond = self.visit(node, node.cond)
        if cond:
            # self.init_scope(node.true)
            self.visit(node, node.true)
            # self.clear_scope(node.true)
        else:
            if node.false is not None:
                # self.init_scope(node.false)
                self.visit(node, node.false)
                # self.clear_scope(node.false)

    def visit_While(self, parent, node):
        cond = self.visit(node, node.cond)
        while cond:
            # self.init_scope(node.block)
            self.visit(node, node.block)
            # self.clear_scope(node.block)
            cond = self.visit(node, node.cond)

    def visit_Repeat(self,parent,node):
        cond=self.visit(node,node.cond)
        while cond:
            self.visit(node, node.block)
            cond=self.visit(node,node.cond)

    def visit_For(self, parent, node):
        self.visit(node, node.init)
        cond = self.visit(node, node.cond)
        while cond:
            #self.init_scope(node.block) #враћено
            self.visit(node, node.block)
            #self.clear_scope(node.block)
            self.visit(node, node.direction)
            cond = self.visit(node, node.cond)

    def visit_Direction(self, parent,node):
        self.visit(node,node.increment)

    def visit_FuncImpl(self, parent, node):
        id_ = self.get_symbol(node.id_)
        id_.params = node.params
        id_.block = node.block
        id_.vars = node.vars
        self.visit(node, node.vars)
        # if node.id_.value == 'main':
        #     self.call_stack.append(node.id_.value)
        #     self.init_scope(node.block)
        #     self.visit(node, node.block)
        #     self.clear_scope(node.block)
        #     self.call_stack.pop()

    def visit_Vars(self, parent, node):
        for v in node.declarations:
            self.visit(parent, v)


    def visit_FuncCall(self, parent, node):
        func = node.id_.value
        args = node.args.args

        # print("FUNC", func)
        if func in ['write', 'writeln']:
            hasArgs=False
            for a in args:
                hasArgs=True
                sym = self.visit(node, a.arg)
                if a.round is not None:
                    # zaokruzi ga na sekond
                    # print(("{0:." + str(a.round.second) + "f}").format(nzm))
                    print(round(sym, a.round.second.value), end='' if func != 'writeln' else '\n')
                elif isinstance(sym, Symbol):
                    print(sym.value, end='' if func != 'writeln' else '\n')
                else:
                    print(sym, end='' if func != 'writeln' else '\n')

            if func=='writeln' and not hasArgs:
                print('')

        elif func == 'readlnn':
            # print("READLN ARGS", args)
            for a in args:
                if not isinstance(a.arg, Id) and not isinstance(a.arg, ArrayElem):
                    raise SystemExit(f"readln argument is not a symbol")
                sym = self.visit(node.args, a.arg)
                inp = input()
                sym.value = self.convert(sym, inp)
        elif func in ['read','readln']:
            if self.linija is None:
                self.linija = input() # "2 3 4"
                self.linijaIndex = 0
            parts = self.linija.split() # ["2", "3", "4"]

            for i, a in enumerate(args):
                sym = self.visit(node.args, a.arg)
                sym.value = parts[self.linijaIndex]
                self.linijaIndex += 1

            if self.linijaIndex == len(parts):
                self.linija = None
                self.linijaIndex = 0



        # elif func in ['read', 'readln']:


            # 2 3 4
            # read(a, b, c)
            # a = 2, b = 3, c = 4

            # 2 3 4
            # readln(a)
            # a = "2 3 4"

            # 234
            # 456
            # readln(a, b)
            # a = "234"
            # b = "456"
            # arguments=[]
            # arguments=args[0].arg.split()
            # for a in arguments

            # for a in args:
            #     if not isinstance(a.arg, Id) and not isinstance(a.arg, ArrayElem):
            #         raise SystemExit(f"readln argument is not a symbol")
            #     sym = self.visit(node.args, a.arg)
            #     sym.value = self.convert(sym, input())


        elif func == 'ord':
            # ord('2') ord(a)
            a = node.args.args[0]
            sym = self.visit(node.args, a.arg); # ovo moze da bude ili symbol pa da ima value ili moze da bude vec value
            if isinstance(sym, Symbol):
                return ord(self.convert(sym, sym.value))
            return ord(sym)
        elif func == 'chr':
            a = node.args.args[0]
            sym = self.visit(node.args, a.arg)

            if isinstance(sym, Symbol):
                return chr(self.convert(sym, sym.value))
            return chr(sym)



        # if func == 'printf':
        #     format_ = args[0].value
        #     format_ = format_.replace('\\n', '\n')
        #     for a in args[1:]:
        #         if isinstance(a, Int):
        #             format_ = format_.replace('%d', a.value, 1)
        #         elif isinstance(a, Char):
        #             format_ = format_.replace('%c', chr(a.value), 1)
        #         elif isinstance(a, String):
        #             format_ = format_.replace('%s', a.value, 1)
        #         elif isinstance(a, Id) or isinstance(a, ArrayElem):
        #             id_ = self.visit(node.args, a)
        #             if hasattr(id_, 'symbols') and id_.type_ == 'char':
        #                 elems = id_.symbols
        #                 ints = [s.value for s in elems]
        #                 non_nulls = [i for i in ints if i is not None]
        #                 chars = [chr(i) for i in non_nulls]
        #                 value = ''.join(chars)
        #             else:
        #                 value = id_.value
        #                 if id_.type_ == 'char':
        #                     value = chr(value)
        #             format_ = re.sub('%[dcs]', str(value), format_, 1)
        #         else:
        #             value = self.visit(node.args, a)
        #             format_ = re.sub('%[dcs]', str(value), format_, 1)
        #     print(format_, end='')
        # elif func == 'scanf':
        #     format_ = args[0].value
        #     inputs = input().split()
        #     matches = re.findall('%[dcs]', format_)
        #     for i, m in enumerate(matches):
        #         id_ = self.visit(node.args, args[i + 1])
        #         if m == '%d':
        #             id_.value = int(inputs[i])
        #         elif m == '%c':
        #             id_.value = ord(inputs[i][0])
        #         elif m == '%s':
        #             word = inputs[i]
        #             length = len(id_.symbols)
        #             for c in word:
        #                 id_.symbols.put(length, id_.type_, None)
        #                 id_.symbols.get(length).value = ord(c)
        #                 length += 1
        # elif func == 'strlen':
        #     a = args[0]
        #     if isinstance(a, String):
        #         return len(a.value)
        #     elif isinstance(a, Id):
        #         id_ = self.visit(node.args, a)
        #         return len(id_.symbols)
        # elif func == 'strcat':
        #     a, b = args[0], args[1]
        #     dest = self.get_symbol(a)
        #     values = []
        #     if isinstance(b, Id):
        #         src = self.get_symbol(b)
        #         elems = [s.value for s in src.symbols]
        #         non_nulls = [c for c in elems if c is not None]
        #         values = [c for c in non_nulls]
        #     elif isinstance(b, String):
        #         values = [ord(c) for c in b.value]
        #     i = len(dest.symbols)
        #     for v in values:
        #         dest.symbols.put(i, dest.type_, None)
        #         dest.symbols.get(i).value = v
        #         i += 1
        else:
            impl = self.global_[func]
            self.call_stack.append(func)
            self.init_scope(impl.block)
            self.visit(node, node.args)
            result = self.visit(node, impl.block)
            self.clear_scope(impl.block)
            self.call_stack.pop()
            self.return_ = False
            return result

    def visit_Block(self, parent, node):
        result = None
        scope = id(node)
        fun = self.call_stack[-1]
        self.scope[fun].append(scope)
        #if len(self.local[scope]) > 5:
        #    exit(0)
        for n in node.nodes:
            if self.return_:
                break
            if isinstance(n, Break):
                break
            elif isinstance(n, Continue):
                continue
            elif isinstance(n, Return):
                self.return_ = True
                if n.expr is not None:
                    result = self.visit(n, n.expr)
            else:
                self.visit(node, n)
        self.scope[fun].pop()
        return result

    def visit_Params(self, parent, node):
        pass

    def visit_Args(self, parent, node):
        print("CALL STACK: ", self.call_stack)

        fun_parent = self.call_stack[-2]
        impl = self.global_[fun_parent]
        self.search_new_call = False
        args = [self.visit(impl.block, a) for a in node.args]
        args = [a.value if isinstance(a, Symbol) else a for a in args]
        fun_child = self.call_stack[-1]
        impl = self.global_[fun_child]
        scope = id(impl.block)
        self.scope[fun_child].append(scope)
        self.search_new_call = True
        for p, a in zip(impl.params.params, args):
            id_ = self.visit(impl.block, p.id_)
            id_.value = a
        self.scope[fun_child].pop()

    def visit_Argument(self, parent, node):
        return self.visit(node, node.arg)

    # def visit_Elems(self, parent, node):
    #     id_ = self.get_symbol(parent.id_)
    #     for i, e in enumerate(node.elems):
    #         value = self.visit(node, e)
    #         id_.symbols.put(i, id_.type_, None)
    #         id_.symbols.get(i).value = value

    def visit_Break(self, parent, node):
        pass

    def visit_Continue(self, parent, node):
        pass

    def visit_Return(self, parent, node):
        pass

    def visit_Type(self, parent, node):
        pass

    def visit_Int(self, parent, node):
        return int(node.value)

    def visit_Char(self, parent, node):
        return str(node.value)

    def visit_String(self, parent, node):
        return str(node.value)

    def visit_Id(self, parent, node):
        return self.get_symbol(node)

    def convert(self, symbol, value):
        if symbol.type_ == 'integer':
            return int(value)
        if symbol.type_ == 'boolean':
            return bool(value)
        if symbol.type_ == 'real':
            return float(value)
        if symbol.type_ == 'string':
            return str(value)
        if symbol.type_ == 'char':
            return str(value)

    def compatible(self, op, first, second):
        # t = {
        #     ["+", "*", "-", "/", "<", ">", "<=", ">="]:     [ ('int', 'int'), ('int', 'float'), ('float', 'float') ], # nije bitan redosled u uredjenoj dvojci
        #     ["mod", "div"]:                                 [ ('int', 'int') ],
        #     ["=", "<>"]:                                    [],
        #     ["and", "or"]:                                  [('bool', 'bool')]
        # }

        if op in ["+", "*", "-", "/", "<", ">", "<=", ">="]:
            return type(first).__name__.lower() in ['int', 'float'] and type(second).__name__.lower() in ['int', 'float'];
        if op in ["mod", "div"]:
            return isinstance(first, int) and isinstance(second, int)
        if op in ["and", "or"]:
            return isinstance(first, bool) and isinstance(second, bool)
        if op in ["=", "<>"]:
            return type(first).__name__ == type(second).__name__



    def visit_BinOp(self, parent, node):
        first = self.visit(node, node.first)
        if isinstance(first, Symbol):
            first = self.convert(first, first.value)
        second = self.visit(node, node.second)
        if isinstance(second, Symbol):
            second = self.convert(second, second.value)


        if node.symbol == '+':
            if not self.compatible('+', first, second):
                raise SystemExit(f"types not compatibile for +")

            return first + second
        elif node.symbol == '-':
            return first - second
        elif node.symbol == '*':
            return first * second
        elif node.symbol == '/':
            return first / second
        elif node.symbol == 'div':
            return first // second
        elif node.symbol == 'mod':
            return first % second
        elif node.symbol == '=':
            return first == second
        elif node.symbol == '<>':
            return first != second
        elif node.symbol == '<':
            return first < second
        elif node.symbol == '>':
            return first > second
        elif node.symbol == '<=':
            return first <= second
        elif node.symbol == '>=':
            return first >= second
        elif node.symbol in ['and', 'AND']: # nzm...
            print("FIRST", first, "SECOND", second)
            return first and second
        elif node.symbol == 'or':
            return first or second
        else:
            return None

    def visit_UnOp(self, parent, node):
        first = self.visit(node, node.first)
        backup_first = first
        if isinstance(first, Symbol):
            first = first.value
        if node.symbol == '-':
            return -first
        elif node.symbol == 'not':
            return not first
        elif node.symbol == '&':
            return backup_first
        else:
            return None

    def run(self):
        self.visit(None, self.ast)
```
